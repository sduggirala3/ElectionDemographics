```python
import pandas as pd
import numpy as np
import sys
np.set_printoptions(threshold=sys.maxsize)
import warnings
pd.set_option("max_columns", 500)
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
```

```python
# import cleaned dataset
data = pd.read_csv('features_v1.csv').drop('Unnamed: 0', 1)
data
```

```python
# high correlation between professional, income, unemployment, and perc_employed
# will use income and unemployment as features, drop other two

data = data.drop(['professional', 'perc_employed'], 1)
data
```

```python
# convert party into binary, drop county and state variables, reorder columns

def party_to_binary(party):
    if party == 'R':
        return 1
    else:
        return 0
    
data['party'] = data['party'].apply(lambda x: party_to_binary(x))
data
```

```python
# get dummy variables for states and concat with dataset
states = data['state']
states_dummies = pd.get_dummies(states)
states_dummies

data = pd.concat([data, states_dummies], axis = 1).drop(['county', 'state'], 1)
data
```

```python
# scale all columns between 0 and 1
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
data_scaled = pd.DataFrame(scaler.fit_transform(data), columns=data.columns)

data_scaled
```

```python
data_scaled = data_scaled[['party','totalpop', 'income', 'service', 'office', 'drive', 'carpool',
       'transit', 'workathome', 'meancommute', 'unemployment',
       'perc_men', 'perc_white', 'perc_private_work', 'perc_citizen', 'AL',
       'AR', 'AZ', 'CA', 'CO', 'CT', 'DC', 'DE', 'FL', 'GA', 'HI', 'IA', 'ID',
       'IL', 'IN', 'KS', 'KY', 'MA', 'MD', 'ME', 'MI', 'MN', 'MO', 'MS', 'MT',
       'NC', 'ND', 'NE', 'NH', 'NJ', 'NM', 'NV', 'NY', 'OH', 'OK', 'OR', 'PA',
       'RI', 'SC', 'SD', 'TN', 'TX', 'UT', 'VA', 'VT', 'WA', 'WI', 'WV', 'WY']]
```

```python
data_scaled.to_csv('features_v3.csv', index = False)
```
